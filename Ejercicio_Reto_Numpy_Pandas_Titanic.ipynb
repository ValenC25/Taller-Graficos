Problema: Análisis de datos meteorológicos. Con Numpy

Supongamos que tenemos un conjunto de datos meteorológicos que contiene la temperatura por meses máxima en grados Celsius registrada en diferentes ciudades (5 ciudades) durante un año. El objetivo es realizar las siguientes operaciones utilizando NumPy:

Asumiendo que los datos están organizados en un array NumPy llamado temperaturas donde cada fila representa un día y cada columna representa una ciudad diferente

- Encontrar la temperatura máxima registrada en todas las ciudades durante el año.
- Calcular la temperatura promedio diaria para cada ciudad.
- Identificar los meses en los que la temperatura superó los 30 grados Celsius en al menos una ciudad.
- Calcular la cantidad de días de lluvia, si se considera un día de lluvia cuando la temperatura máxima es inferior a 10 grados Celsius.

Las temperaturas son las siguientes:

```python
import numpy as np
temperaturas =np.array([[26, 28, 30, 29, 27],
              [31, 32, 33, 30, 29],
              [29, 27, 28, 31, 30],
              [32,35,36,23,32],
               [8,12,15,4,8],
               [36,39,25,28,32],
               [23,24,25,35,32],
               [25,34,28,25,22],
               [12,11,14,25,15],
               [26,27,30,31,29],
               [12,8,12,6,9],
               [25,21,27,21,28]])

#temperatura máxima por ciudad
print (np.max (temperaturas, axis = 0))

#temperatura promedio por cada ciudad
print(np.mean(temperaturas, axis = 0))

#mayor a 30 grados celsius
print('La cantidad de meses que supero los 30 grados: ',np.sum(np.sum(temperaturas > 30, axis = 1) >= 1))

#menor a 10 grados celsius, lluvia
print('la cantidad de meses en los que hubo lluvia: ',np.sum(np.sum(temperaturas < 10, axis = 1) >= 1))
```

```python
titanic_data.groupby(['Sexo' , 'Sobrevivente', 'Clase'])['ID'].count()
```

```python
#8 agrupar por sexo y clase para determinar cuantos hombre y mujeres hay en cada clase

titanic_data.groupby(['Clase', 'Sexo'])['ID'].count()

#9 Calcular el porcentaje de pasajeros que sobrevivieron en cada clase

mi_series = titanic_data.groupby(['Sobrevivente','Clase'])['ID'].count() # muestra la respuesta horizontal
print(mi_series.values * 100 / mi_series.values.sum())
mi_series

```

```python
import numpy as np

mi_series = titanic_data.groupby(['Sobrevivente','Clase'])['ID'].count()
porcentaje = mi_series.values * 100 / mi_series.values.sum()
mi_series = mi_series.reset_index() # lo combierte una serie en  dataframe,
mi_series['Tasa de supervivientes en %'] = porcentaje # muestra el número completo
mi_series['Tasa de supervivientes en %'] = np.round(porcentaje,3) # lo redondea a 3 cifras significativas
mi_series
```
